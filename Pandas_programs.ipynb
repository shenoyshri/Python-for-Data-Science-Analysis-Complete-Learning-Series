## Compare DataFrames

```python
dict = {"Fruits": ["mango","apples","banana","papaya"],
       "Price": [100,150,50,35],
       "Quantity":[15,10,10,3]}
df1 = pd.DataFrame(dict)
print(df1)
df2 = df1.copy()

df2.loc[0,"Price"] = 120
df2.loc[1,"Price"] = 175
df2.loc[3,"Price"] = 30
df2.loc[0,"Quantity"] = 12
df2.loc[1,"Quantity"] = 15
df2.loc[3,"Quantity"] = 5
print(df2)

print(df1.compare(df2,align_axis = 0))
```

```python
print(df1.compare(df2))
```

```python
print(df1.compare(df2,keep_equal= False))
```

```python
print(df1.compare(df2,keep_shape= True))
```

## Pandas Pivoting and Melting DataFrames

```python
dict= {"Keys": ["k1","k2","k1","k2"],
      "Names": ["John","Ben","David","Peter"],
      "Houses": ["red","blue","green","red"],
      "Grades": ["3rd","8th","9th","8th"]}
df = pd.DataFrame(dict)
print(df)
df.pivot(index="Keys", columns="Names", values=["Houses","Grades"])
```

### Melting

```python
dict= {"Names": ["John","Ben","David","Peter"],
      "Houses": ["red","blue","green","red"],
      "Grades": ["3rd","8th","9th","8th"]}
df= pd.DataFrame(dict)
print(df)
```

```python
print(df.melt("Names",["Houses","Grades"],"House&Grades"))
```
