```python
# Write a function that takes a number as a parameter ans check if the number is prime or not
def check_prime(num):
    if num == 1:
        print("it is not a prime number")
    if num >=2:
        for i in range(2,num):
            if num % 2 == 0:
                print("it is a not a prime number")
                break
        else:
            print("it is a prime number")

check_prime(29)
        
```

```python
# WA python function to summ all the numbers in a list
def sum_all(num):
    total = 0
    for i in num:
        total+=i
    return(total)
    
print("The sum of all the items in the list is: ", sum_all([12,4,5,6,7,8]))
```

```python
# Solution 2 using Recursion
def add_num(num):
    if len(num) == 1:
        return(num[0])
    else:
        return (num[0]) + add_num(num[1:])

print("The sum of all the items in the list is: ", add_num([12,4,5,6,7,8]))
```

```python
# WAP to find the nth element of the fibonacci series
def fibs(num):
    if num == 1:
        return(0)
    elif num == 2:
        return(1)
    else:
        return(fibs(num-1) + fibs(num-2))

n = int(input('enter a number:'))
print(f"The {n}th element of fibonaccci series is ", fibs(n))
```
