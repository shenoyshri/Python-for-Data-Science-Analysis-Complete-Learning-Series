```python
import numpy as np
# Addition
arr1 = np.array([[30,40],[20,10]])
arr2 = np.array([[30,20],[10,20]])
print(arr1 + arr2)
print(np.add(arr1,arr2))
```

```python
# Subtraction
arr1 = np.array([[30,40],[20,10]])
arr2 = np.array([[30,20],[10,20]])

print(np.subtract(arr1,arr2))
```

```python
# Multiplication
arr1 = np.array([[30,40],[20,10]])
arr2 = np.array([[30,20],[10,20]])

print(arr1 * arr2)
print(np.multiply(arr1,arr2))
```

```python
# Division

arr1 = np.array([[30,40],[20,10]])
arr2 = np.array([[30,20],[10,20]])

print(arr1/arr2)
print(np.divide(arr1,arr2))
```

```python
arr1 = ([2,3,4,5,6])
arr2 = ([2])

print("The square of number is",np.pow(arr1,arr2))
```

```python
arr1 = np.array([4,9,16,25,36])

print("The square root of the no is", np.sqrt(arr1))
```
